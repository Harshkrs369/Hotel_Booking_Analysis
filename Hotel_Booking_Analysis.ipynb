# Data Analysis and Visualization

```python
cancelled_perc = (df['is_canceled'].value_counts(normalize=True))*100
cancelled_perc
```

```python
plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),color='orange',edgecolor='k',width=0.4)
plt.show()
```

*The accompanying bar graph show the count of reservations that are cancelled. There is 37% of clients who cancelled their reservations, which has significant impact on hotel earnings.*

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='gist_earth',width=0.4)
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('Hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

*In comparison of resort hotel, city hotel has more bookings. It's possible that resort hotels are more expensive than city hotels.*

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
(resort_hotel['is_canceled'].value_counts(normalize=True))*100
```

```python
city_hotel=df[df['hotel']=='City Hotel']
(city_hotel['is_canceled'].value_counts(normalize=True))*100
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
```

*The line graph shows that on certain days, the average daily rate for city hotel is less than that of a resort hotel, and on other days it is even less. It goes without saying that weekends and holidays may see rise in resort hotel rates.*

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='colorblind')
for bars in ax1.containers:
    ax1.bar_label(bars)
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservations per month',size=20)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

*We developed a grouped bar graph to analyse months with highest and lowest reservations levels according to reservation status. The number of confirmed reservation is highest in August whereas January is the month with most cancelled reservations.*

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month', fontsize=30)
ax1=sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

*This bar graph shows the average daily rate of each month. It demonstrates that the cancellation are most common when the prices are greatest and least common when the prices are lowest. Therefore, the cost of accomodation is solely responsible for the cancellation.*

```python
cancelled_data = df[df['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation cancelled',fontsize=30)
patches, texts, autotexts = plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)

# Set data label text color to black
for autotext in autotexts:
    autotext.set_color('black')
plt.show()
```

*The above pie chart demonstrates the highest reservation cancelled of top 10 countries. The top country is portugal with highest number of cancellation.*

```python
cdadr= cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cdadr.reset_index(inplace=True)
cdadr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data = df[df['is_canceled']==0]
ncdadr= not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
ncdadr.reset_index(inplace=True)
ncdadr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(ncdadr['reservation_status_date'],ncdadr['adr'],label='not_cancelled')
plt.plot(cdadr['reservation_status_date'],cdadr['adr'],label='cancelled',alpha=0.7)
plt.legend()
```

```python
cdadr=cdadr[(cdadr['reservation_status_date']>'2016') & (cdadr['reservation_status_date']<'2017-09')]
ncdadr=ncdadr[(ncdadr['reservation_status_date']>'2016') & (ncdadr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.scatter(ncdadr['reservation_status_date'],ncdadr['adr'],label='not_cancelled',marker='*')
plt.scatter(cdadr['reservation_status_date'],cdadr['adr'],label='cancelled',marker='*',alpha=0.6)
plt.legend(fontsize=20)
```

*As seen in the scatter plot, reservations are cancelled when average daily rate is higher than when it is not cancelled. It clearly proves all the above analysis that higher price leads to higher cancellation.*

```python
# Plot the distribution of lead time
plt.figure(figsize=(12, 6))
sns.histplot(df['lead_time'], bins=25, kde=True)
plt.title('Distribution of Lead Time for Bookings')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.show()

# Boxplot of lead time by hotel type
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='hotel', y='lead_time')
plt.title('Lead Time by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Lead Time (days)')
plt.show()
```

*In the above hist and box plot it can be inferred that the majority of the bookings are made 0-100 days before check-in. It can be said customers often wait for last minute deals and discounts.*

```python
# Plotting distribution of lead time for canceled and non-canceled bookings
plt.figure(figsize=(14, 6))

# Distribution of lead time for non-canceled bookings
plt.subplot(1, 2, 1)
sns.histplot(df[df['is_canceled'] == 0]['lead_time'], bins=25, kde=True, color='green')
plt.title('Lead Time for Non-Canceled Bookings')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')

# Distribution of lead time for canceled bookings
plt.subplot(1, 2, 2)
sns.histplot(df[df['is_canceled'] == 1]['lead_time'], bins=25, kde=True, color='red')
plt.title('Lead Time for Canceled Bookings')
plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

*From the above histplots we can say that the last minute cancellation rate is high. It may be due to various factors among which linient cancellation policies offered by hotels can play major role in it. Poor hotel services, overbooking and better offers elsewhere also can be the possible reasons for last minute cancellations.*
